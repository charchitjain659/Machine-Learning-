# Question 1 : What is Dimensionality Reduction? Why is it important in machine learning?
# answer
Dimensionality reduction is the process of reducing the number of input variables (features) in a dataset while preserving important information.
It helps simplify models, remove noise, reduce computation time, and prevent overfitting, especially when dealing with high-dimensional data.

** Example: ** Using PCA to reduce 100 features to 10 key components that explain most of the data variance.

# Question 2: Name and briefly describe three common dimensionality reduction techniques..
# answer
***1. Principal Component Analysis (PCA): ***Transforms features into new uncorrelated variables (principal components) that capture maximum variance.

**2. Linear Discriminant Analysis (LDA):** Finds linear combinations of features that best separate two or more classes.

**3. t-Distributed Stochastic Neighbor Embedding (t-SNE)**: A non-linear technique used mainly for visualizing high-dimensional data in 2D or 3D space.

# Question 4: Explain the concept of anomaly detection and its significance
# answer
Anomaly detection identifies rare or unusual data points that deviate significantly from the normal pattern.
It is significant because it helps detect fraud, network intrusions, machine failures, or unusual customer behavior early.

# Question 5: List and briefly describe three types of anomaly detection techniques.
# Answer:
**1 Statistical Methods:** Identify outliers based on statistical properties (e.g., z-score, IQR).

**2 Machine Learning Methods:** Use models like Isolation Forest, One-Class SVM, or Autoencoders to detect anomalies.

**3 Distance-Based Methods:** Detect anomalies based on their distance from other points (e.g., KNN-based outlier detection).

# Question 6: What is time series analysis? Mention two key components of time series data.
# Answer:
Time series analysis studies data points collected or recorded over time to identify trends, patterns, and seasonal effects.
**1 Trend:** The long-term movement in data over time.
**2 Seasonality:** Regular repeating patterns at fixed time intervals (e.g., monthly or yearly).

# Question 8: Write Python code to perform K-means clustering on a sample dataset.
# answer


```python
# Import libraries
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Create a sample dataset
X, y = make_blobs(n_samples=200, centers=3, random_state=42)

# Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', marker='x', s=200, label='Centroids')
plt.title("K-Means Clustering Example")
plt.legend()
plt.show()
```

# Question 9: What is inheritance in OOP? Provide a simple example in Python.
# answer
Inheritance allows a class (child/subclass) to acquire properties and methods from another class (parent/superclass). It promotes code reusability and hierarchical structure.

```python
class Animal:
    def sound(self):
        print("Animals make sounds")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

# Create object
d = Dog()
d.sound()  # Output: Dog barks
```
