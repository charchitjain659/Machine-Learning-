# Question 1: Explain the differences between AI, ML, Deep Learning (DL), and Data Science (DS).
# answer
1. **Artificial Intelligence (AI)**
***Definition:*** AI is the broad field of computer science focused on creating systems that can perform tasks that normally require human intelligence.
***Goal: ***Make machines "think" and "act" intelligently.
***Scope:*** Includes reasoning, problem-solving, speech recognition, language understanding, and decision-making.
***Example:*** Chatbots, self-driving cars, and smart assistants like Siri or Alexa.

** 2. Machine Learning (ML)**
***definition:*** ML is a subset of AI that allows machines to learn from data and improve their performance over time without explicit programming.
***Goal:*** Enable systems to automatically improve based on experience.
***Techniques:*** Supervised learning, unsupervised learning, reinforcement learning.
***Example: ***Email spam filters, recommendation systems, fraud detection.

**3. Deep Learning (DL) **
***Definition:*** DL is a specialized subset of ML that uses neural networks with many layers (deep neural networks) to model complex patterns in data.
***Goal:*** Automatically learn hierarchical features from raw data (images, text, audio).
***Requirements:*** Large amounts of data and high computational power.
***Example:*** Image recognition, voice assistants, autonomous vehicles’ vision systems.

**4. Data Science (DS)**
***Definition:*** DS is an interdisciplinary field that uses statistical analysis, ML, and data visualization to extract insights from structured and unstructured data.
***Goal: ***Make data-driven decisions and predictions.
***Techniques: ***Data cleaning, exploratory data analysis, ML/DL modeling, visualization.
***Example:*** Market trend analysis, customer segmentation, predictive maintenance.


# Question 2: What are the types of machine learning? Describe each with one real-world example.
# answer
**1. Supervised Learning**
***defination*** supervised learning, the model is trained on a labeled dataset—each input has a corresponding output. The model learns to map inputs to outputs.
***Real-World Example:***
Email spam detection: The model is trained with emails labeled as “spam” or “not spam” and predicts the label for new emails.

**2. Unsupervised Learning**
***Definition: ***In unsupervised learning, the data does not have labels. The model finds patterns, relationships, or structures in the data on its own.
**`Real-World Example:`**
Customer segmentation: E-commerce companies group customers based on buying behavior without predefined categories.

**3. Semi-Supervised Learning**
***Definition:*** Combines both labeled and unlabeled data. The model is trained with a small amount of labeled data and a large amount of unlabeled data.
***Real-World Example:***
Medical diagnosis: A few patient records are labeled, but many unlabeled records are used to train models for disease prediction.

**4. Reinforcement Learning (RL)**
***Definition: ***The model (agent) learns by interacting with an environment and receiving rewards or penalties for actions. It aims to maximize cumulative reward.
***Real-World Example:***
Self-driving cars: The car learns to navigate roads safely by receiving rewards for good driving actions and penalties for mistakes.

# Question 3: Define overfitting, underfitting, and the bias-variance tradeoff in machine learning.
# answer
***1. Overfitting***

***Definition: ***
 Overfitting occurs when a machine learning model learns the training data too well, including noise or random fluctuations, and performs poorly on new, unseen data.
***Characteristics:***
Very low error on training data.
High error on test/validation data.
***Cause:*** Model is too complex (too many parameters) for the amount of data.
***Example:  ***A decision tree that perfectly classifies training data but fails on new examples.

***2. Underfitting***

***Definition***: Underfitting occurs when a model cannot capture the underlying patterns in the data, resulting in poor performance on both training and test data.
***Characteristics***:
High error on training data.
High error on test/validation data.
***Cause***: Model is too simple or insufficient training.
**Example**: A linear model trying to fit highly nonlinear data.

***3. Bias-Variance Tradeoff***
***Definition***: The bias-variance tradeoff is the balance between a model’s ability to generalize (low variance) and its accuracy on training data (low bias).
***Key Points:***
Bias: Error due to overly simplistic assumptions in the model (causes underfitting).
Variance: Error due to model being too sensitive to training data (causes overfitting).
***Goal:*** Find a sweet spot where both bias and variance are minimized for best generalization.
***Illustration***:
High bias + low variance → underfitting.
Low bias + high variance → overfitting.
Moderate bias + moderate variance → optimal model.

# Question 4: What are outliers in a dataset, and list three common techniques for handling them.
# answer
Outliers are data points that are significantly different from the majority of the data in a dataset. They deviate markedly from other observations and may occur due to variability in the data, errors in measurement, or data entry mistakes. Outliers can distort statistical analyses, affect model performance, and lead to misleading results if not handled properly.
***Three common techniques for handling outliers:***
***Removal (Deletion):***
If an outlier is clearly due to an error or is not relevant to the analysis, it can be removed from the dataset.
***Transformation:***
Apply mathematical transformations to reduce the impact of outliers. Common transformations include logarithmic, square root, or Box-Cox transformations.
*** Imputation / Capping (Winsorization):***
Replace outliers with a threshold value, such as the nearest percentile (e.g., 1st or 99th percentile), or use statistical measures like mean/median.



# Question 5: Explain the process of handling missing values and mention one imputation technique for numerical and one for categorical data
# answer
Missing values occur when some data points are not recorded or are lost in a dataset. Proper handling is crucial because missing data can bias analyses, reduce statistical power, and affect machine learning model performance. The general process involves the following steps:
***1. Identify Missing Values:***
Check for null, NaN, or blank entries in the dataset.
***2. Analyze Missing Patterns:***
Determine if missingness is random (MCAR), depends on observed data (MAR), or depends on unobserved data (MNAR).
This helps decide the right strategy to handle missing values.
***3. Handle Missing Values:***
Removal: Delete rows or columns with missing data (used if missing data is small or not critical).
Imputation: Replace missing values with estimated values.

# Question 8: Write a Python script to: ● Generate 1000 samples from a normal distribution. ● Introduce 50 random missing values. ● Fill missing values with the column mean. ● Plot a histogram before and after imputation.
# answer

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Generate 1000 samples from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)  # mean=50, std=10

# Convert to pandas DataFrame
df = pd.DataFrame(data, columns=['Value'])

# Step 2: Introduce 50 random missing values
missing_indices = np.random.choice(df.index, size=50, replace=False)
df.loc[missing_indices, 'Value'] = np.nan

# Plot histogram before imputation
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.hist(df['Value'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram Before Imputation')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Step 3: Fill missing values with column mean
df['Value'].fillna(df['Value'].mean(), inplace=True)

# Plot histogram after imputation
plt.subplot(1,2,2)
plt.hist(df['Value'], bins=30, color='lightgreen', edgecolor='black')
plt.title('Histogram After Imputation')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

# Question 10: You are working as a data scientist for a retail company. You receive a customer transaction dataset that contains: ● Missing ages ● Outliers in transaction amount, ● A highly imbalanced target (fraud vs. non-fraud), ● Categorical variables like payment method. Explain the step-by-step data preparation plan you’d follow before training a machine learning model. Include how you’d address missing data, outliers, imbalance, and encoding.
# answer
**1. Understand the Dataset**
Examine data types, ranges, and summary statistics using df.info(), df.describe().
Identify missing values, outliers, class imbalance, and categorical variables.
**2. Handle Missing Data**
Missing Ages (numerical column):
If missing values are random, impute using median (robust to outliers) or mean.
** 3. Handle Outliers**
Transaction Amount:
Detect outliers using IQR method or Z-score.
Options to handle:
Cap/Floor (Winsorization): Replace extreme values with 1st and 99th percentile.
Log transformation: Reduce skewness if values vary widely.
** 4. Handle Imbalanced Target**
Fraud vs. Non-Fraud (highly imbalanced):
Use resampling techniques:
Oversampling: e.g., SMOTE (Synthetic Minority Oversampling Technique) to generate synthetic minority samples.
Undersampling: Reduce majority class to balance dataset.
Choose technique depending on dataset size and model type.
**5. Encode Categorical Variables**
Payment Method (categorical):
If nominal (no order): use One-Hot Encoding.
example: pd.get_dummies(df['PaymentMethod'])
If ordinal: use Label Encoding.
Ensure encoding does not introduce bias.
** 6. Feature Scaling**
Scale numerical features (e.g., Age, Transaction Amount) using:
Min-Max Scaling or Standardization (Z-score), especially for algorithms sensitive to scale (SVM, KNN, neural networks).
**7. Split Data**
split dataset into training and test sets (e.g., 80-20) before resampling to avoid data leakage.
**8. Final Checks **
Verify no missing values remain.
Check that class distribution is appropriate in the training set.
Ensure features are properly encoded and scaled.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

# -------------------------------
# Step 1: Create a sample dataset
# -------------------------------
# Increased the number of samples to provide enough data for SMOTE
n_samples = 1000 # Total samples
n_fraud = 100 # Number of fraud samples (minority class)
n_non_fraud = n_samples - n_fraud # Number of non-fraud samples (majority class)

data = {
    'Age': np.random.randint(20, 70, n_samples).tolist(),
    'TransactionAmount': np.random.uniform(50, 5000, n_samples).tolist(),
    'PaymentMethod': np.random.choice(['Card', 'Cash', 'UPI'], n_samples).tolist(),
    'Fraud': [0]*n_non_fraud + [1]*n_fraud # Imbalanced target
}

df = pd.DataFrame(data)

# Introduce missing values and outliers after DataFrame creation
missing_indices_age = np.random.choice(df.index, size=int(0.1*n_samples), replace=False) # 10% missing age
df.loc[missing_indices_age, 'Age'] = np.nan

outlier_indices_amount = np.random.choice(df.index, size=int(0.01*n_samples), replace=False) # 1% outliers in amount
df.loc[outlier_indices_amount, 'TransactionAmount'] = df.loc[outlier_indices_amount, 'TransactionAmount'] * 100 # Introduce outliers

# -------------------------------
# Step 2: Handle Missing Values
# -------------------------------
df['Age'].fillna(df['Age'].median(), inplace=True)

# -------------------------------
# Step 3: Handle Outliers
# -------------------------------
Q1 = df['TransactionAmount'].quantile(0.01)
Q99 = df['TransactionAmount'].quantile(0.99)
df['TransactionAmount'] = df['TransactionAmount'].clip(Q1, Q99)

# -------------------------------
# Step 4: Encode Categorical Variables
# -------------------------------
df = pd.get_dummies(df, columns=['PaymentMethod'], drop_first=True)

# -------------------------------
# Step 5: Scale Numerical Features
# -------------------------------
scaler = MinMaxScaler()
df[['Age', 'TransactionAmount']] = scaler.fit_transform(df[['Age', 'TransactionAmount']])

# -------------------------------
# Step 6: Handle Imbalanced Target
# -------------------------------
X = df.drop('Fraud', axis=1)
y = df['Fraud']

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

# -------------------------------
# Step 7: Train-Test Split
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)

# -------------------------------
# Final Output
# -------------------------------
print("Processed Feature Data (X_train):")
display(X_train.head())
print("\nResampled Target Data (y_train):")
print(y_train.value_counts())
```
